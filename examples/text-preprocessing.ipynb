# Text Manipulation 
- Removal Stopwords
- Word Normalization
- Stem or Lemmatization

### Removal Stopwords

```python
# Example
texts_ = [
    "Please submit the report ASAP",
    "FYI, the new software update will be available next Monday",
]
texts_ = word_tokenize(texts_)
stop_words = ["the","Please"]
remove_stopwords(texts_,lang='english',stopwords=stop_words,return_token=False)
```

If you don't fill in the stopwords parameter, it will automatically take the following stopwords by default :

```python
get_stopwords_en()
```

### Word Normalization

```python
# Example
norm_words = {"ASAP": "as soon as possible", "FYI": "For Your Information"}
texts_ = [
    "Please submit the report ASAP",
    "FYI, the new software update will be available next Monday",
]
texts_ = word_tokenize(texts_)
normalize(tokens=texts_, norm_words=norm_words,return_token=False)
```

### Stem

```python
# Example
texts_ = ["She is reading a book.",
"They were running in the park.",
"The cats are sitting on the mat.",
"He studies mathematics every day."
]
texts_ = word_tokenize(texts_)
stemming(data=texts_,lang='english',return_token=False,verbose=False)
```

### Lemmatization

```python
# Example
lemmatization(data=texts_,lang="english",return_token=False,verbose=False)
```
